### Descenso de gradiente

```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
f = lambda X: X[0]**2 + X[1]**2  #FUNCIÓN x^2 + y^2
df = lambda X: [2*X[0], 2*X[1]]  #GRADIENTE (la derivada)

#preparo los datos para dibujar
resolucion=30
rango=2.5
X=np.linspace(-rango, rango, resolucion)
Y=np.linspace(-rango, rango, resolucion)
Z=np.zeros([resolucion, resolucion])

for ix, x in enumerate(X):
    for iy,y in enumerate(Y):
        Z[iy,ix] = f([x,y])

#Pinto el mapa de niveles Z
contour=plt.contourf(X,Y,Z, resolucion)
cbar=plt.colorbar(contour)

#generamos punto de partida aleatorio
P=[random.uniform(-rango, rango), random.uniform(-rango, rango)]
print(P)
plt.plot(P[0], P[1], "o", c="blue")
#tasa aprendizaje
TA=0.05

#ITERACIONES DEL ALGORITMO DE DESCENSO DE GRADIENTE
for _ in range(500):
    grad=df(P)
    P[0], P[1] = P[0] - TA*grad[0], P[1] - TA*grad[1]
    plt.plot(P[0], P[1], "o", c="red", markersize=0.4)
    
plt.plot(P[0], P[1], "o", c="yellow")
print("Solución:", P, f(P))
```
